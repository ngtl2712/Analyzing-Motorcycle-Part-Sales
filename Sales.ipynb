2. Find out how much Wholesale net revenue each product_line generated per month per warehouse in the dataset.

```python
%%sql
SELECT
    product_line,
    CASE
        WHEN DATE_PART('month', date::DATE) = 6 THEN 'June'
        WHEN DATE_PART('month', date::DATE) = 7 THEN 'July'
        WHEN DATE_PART('month', date::DATE) = 8 THEN 'August'
    END AS month,
    warehouse,
    SUM(total) - SUM(payment_fee) AS net_revenue
FROM sales
WHERE client_type = 'Wholesale'
GROUP BY product_line, 
         month,
         warehouse
ORDER BY product_line, 
         month,
         net_revenue DESC;
```
